## Popular Windows PowerShell+CMD Prompt Commands 

### windows+s=>search `cmd`

<code>ipconfig</code> : ipconfig : brings up your system's IP address information

<code>ipconfig /all</code>

<code>ipconfig /all | findstr DNS</code> : Filter and search for DNS information 

<code>ipconfig /release</code> Then <code>ipconfig /renew</code> : Refreshes your IP Address information and all interfaces

<code>ipconfig /displaydns</code> : ALL IP addresses

<code>ipconfig /displaydns | clip</code> : copy all that website visited inforamtion to the clipboard

<code>ipconfig /flushdns</code> : Flush Your DNS, renew your stale DNS servers 

<code>nslookup -type=ptr</code>

<code>cls</code> : clear your screen

<code>getmac /v</code> : get all your MAC addresses

<code>powercfg /energy</code>

<code>powercfg /batteryreport</code>

<code>assoc</code> : show your file types are associated with which programs 

<code>assoc .mp4=VLC.vlc</code> : make the .mp4 files open with .vlc

<code>chkdsk /f</code> : /f go through your system to see if there are any issues/errors to fix them

<code>chkdsk /r</code> : /r go through your system and check physical sector issues 

<code>sfc /scannow</code> : system file checker 

<code>DISM /Online /Cleanup /CheckHealth</code> L:fixes your system's image 

<code>DISM /Online /Cleanup /ScanHealth</code> L:fixes your system's image; deeper check

<code>DISM /Online /Cleanup /RestoreHealth</code> L:fixes your system's image; help to not re-install your windows operating system

<code>tasklist | findstr script</code>

<code>taskkill /f /pid script_##</code>

<code>netsh wlan show wlanreport</code>

<code>netsh interface show interface</code> : showing you your interfaces

<code>netsh interface ip show address | findstr "IP Address"</code>

<code>netsh interface ip show dnsservers</code>

<code>netsh firewall set allprofiles state off</code>

<code>netsh firewall set allprofiles state on</code>

<code>ping -t website</code> : -t continuous

<code>tracert networkchuck.com</code> : how long to get to that website and route

<code>tracert -d networkchuck.com</code>

<code>netstat -af</code> : what ports are open, and status for your connections 

<code>netstat -o</code> : processid shown for all your connections 

<code>netstat -e -t 5</code> : statistics for every 5 seconds 

<code>route print</code> : routes you will take to get to specific gateways

<code>route add IPADDRESS MASK GETWAY</code>

<code>shutdown</code>

<code>shutdown /r /fw /f /t 0</code> :shutdown your computer and go into the system bios
